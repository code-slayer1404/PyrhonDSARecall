```python
import random, collections
class TreeNode():
    def __init__(self,val,left=None,right=None):
        self.val = val
        self.left = left
        self.right = right
    
    def __str__(self):
        return str(self.val)
    
    @staticmethod
    def generate(size):
        arr = collections.deque()
        for _ in range(size):
            arr.append(random.randint(-9,9))
        
        root = TreeNode(arr.popleft())
        
        for num in arr:
            root = TreeNode.insert(root,num)
        
        return root
    
    @staticmethod
    def insert(node,val):
        newNode = TreeNode(val)

        if node is None:
            return newNode
        
        head = node
        while node:
    
            if val < node.val:
                if node.left is None:
                    node.left = newNode
                    break
                node = node.left
            else:
                if node.right is None:
                    node.right = newNode
                    break
                node = node.right
        return head
    
    @staticmethod
    def display(node,step=0):
        if node is None:
            return
        TreeNode.display(node.left,step+1)
        if step ==0:
            print(node)
        else:
            print("     "*(step-1)+"|--->",node.val,sep="")
        
        TreeNode.display(node.right,step+1)
        

node = TreeNode.generate(10)

TreeNode.display(node)
```

```python
def houseRobber(node):
    if node is None:
        return (float("-inf"),0) # penalizing robbing a null node

    leftRobbed,leftNotRobbed = houseRobber(node.left)
    rightRobbed,rightNotRobbed = houseRobber(node.right)
    
    nodeRobbed = node.val + (leftNotRobbed+rightNotRobbed)
    nodeNotRobbed = max(leftRobbed,leftNotRobbed)+max(rightRobbed,rightNotRobbed)
    return (nodeRobbed,nodeNotRobbed)
```

```python
def cameras(node):
    if node is None:
        return (float("inf"),0,0) # state is has camera, covered by either child , not covered by children or itself
    # (hasCam, coveredByChild, needsParent) is a better naming

    
    leftHas,leftCovered,leftNotCovered = cameras(node.left)
    rightHas,rightCovered,rightNotCovered = cameras(node.right)
    
    nodeHas = 1 + min( leftHas,leftCovered,leftNotCovered) + min(rightHas,rightCovered,rightNotCovered)
    nodeCovered = min(leftHas+min(rightHas,rightCovered),rightHas+ min(leftHas,leftCovered))
    nodeNotCovered = rightCovered + leftCovered # parent will handle
    return (nodeHas,nodeCovered,nodeNotCovered)
```
