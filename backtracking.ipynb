```python

# sort of unbounded knapsack
def bt(arr,k):
    
    stack = [(0,[])]
    
    while  stack:
        i,path = stack.pop()
        if len(path)==k:
            for _ in path:
                print(arr[_],end=" ")
            print()
            continue
        
        for j in range(i,len(arr)):
            stack.append((j,path+[j]))


bt([1,2,3],2)

print("\n\n")
# sort of 0/1 knapsack
def bt(arr,k):
    
    stack = [(0,[])]
    
    while  stack:
        i,path = stack.pop()
        if len(path)==k:
            for _ in path:
                print(arr[_],end=" ")
            print()
            continue
        
        for j in range(i,len(arr)):
            stack.append((j+1,path+[j]))


bt([1,2,3],2)
```

```python

#permutations
def bt(arr):
    
    stack = [([])]
    
    while  stack:
        path = stack.pop()
        if len(path)==len(arr):
            for _ in path:
                print(arr[_],end=" ")
            print()
            continue
    
        for j in range(len(arr)):
            if j in path:
                continue
            stack.append((path+[j]))


bt([1,2,3,4,5])
```

```python
        
# non bt approach.
# we find a pivot index from the right hand side from where the array displays ascending behaviour
# now we have pivot and terms right of pivot are in descending order.
# the descending order part is max in numerical value so it makes sense that pivot must be diff
# 2 5 4 3 : nothing I can do in 5 4 3. So one of these (smallest i.e. 3) will have to replace pivot
# we select the smallest term greater than pivot(from right to handle duplicates better), temp
# we swap pivot and temp
# now the part is greater than original but not smaleest greater as part from index pivot+1 to end is in descending
# simply reverse this part and voilla!
def nextPermutation(arr:list):
    
    pivot = -1
    for _ in range(len(arr)-2,-1,-1):
        if arr[_] < arr[_+1]:
            pivot = _
            break
    if pivot == -1:
        arr.reverse()
        return
    
    temp = pivot      
    for _ in range(len(arr)-1,pivot,-1):
        if arr[_] > arr[pivot]:
            temp = _
            break
    
    arr[pivot],arr[temp] = arr[temp],arr[pivot]
    
    s,e = pivot + 1, len(arr)-1
    
    while (s<e):
        arr[s],arr[e] = arr[e],arr[s]
        s += 1
        e -= 1
    
    return arr


arr = [1,2,3,4,5]
print(nextPermutation(arr))
print(nextPermutation(arr))
print(nextPermutation(arr))
print(nextPermutation(arr))
print(nextPermutation(arr))
print(nextPermutation(arr))
print(nextPermutation(arr))


# arr = [5,4,3,1,2]
# print(nextPermutation(arr))
```
