kadane problems have two equivalent representations:
1) running sum + reset (often the standard textbook but less intuitive for some variations)
2) prefix sum based version maintaining a prev min so far. More intuitive and easily adaptable

```python
def stocks1(arr):
    n = len(arr)
    minSoFar = arr[0]
    ans = float("-inf")
    for _ in range(1,n):
        temp = arr[_] - minSoFar
        ans = max(ans,temp)
        minSoFar = min(minSoFar,arr[_])
    
    return ans

prices = [7,1,5,3,6,4]
print(stocks1(prices))

prices = [7,1,5,3,6,4,10,20,40,80]
print(stocks1(prices))
```

```python
# comes from diff array concept and hidden prefix sum. Not intuitive directly
def stocks(arr):
    total = 0
    best = 0

    for i in range(1, len(arr)):
        # total = max(0, total + arr[i] - arr[i-1])
        total = max(arr[i]-arr[i-1], total + arr[i] - arr[i-1])
        best = max(best, total)

    return best


prices = [7,1,5,3,6,4,10,20,40,80]
print(stocks(prices))
```

```python
def maxSub(arr):
    maxSumSoFar = 0
    best = float("-inf")
    for i in range(len(arr)):
        maxSumSoFar = max(maxSumSoFar+arr[i],arr[i])
        best = max(best,maxSumSoFar)
    return best

print(maxSub([-4,2,5,1,-3,5]))
print(maxSub([-4,-2,-5,-1,-3,-5]))
```

```python
def maxSub(arr):
    minPrefix = 0 #p[0]
    prefix = 0
    ans = float("-inf")
    for i in range(len(arr)):
        prefix += arr[i] # p[r]
        ans = max(ans,prefix-minPrefix)
        minPrefix = min(minPrefix,prefix)
    return ans

print(maxSub([-4,2,5,1,-3,5]))
print(maxSub([-4,-2,-5,-1,-3,-5]))
```
