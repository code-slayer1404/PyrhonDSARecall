```python
def climb(n,memo=None):
    
    if n < 0:
        return 0
    if n == 0:
        return 1
    if memo is None:
        memo = {}
    
    if n in memo:
        return memo[n]
    
    memo[n] = climb(n-1,memo) + climb(n-2,memo)
    return memo[n]


print(climb(0))
print(climb(1))
print(climb(2))
print(climb(3))
print(climb(4))
    
    
```

```python
def jumpGame2D(matrix,r,p):
    if r == len(matrix):
        return 0
    ans = 0
    for i in range(len(matrix[0])):
        if r >0 :
            ans = max(ans,matrix[r][i] - abs(i-p) + jumpGame2D(matrix,r+1,i))
        else:
            ans = max(ans,matrix[r][i] + jumpGame2D(matrix,r+1,i))
    
    return ans
points = [
        [1,0,3,4],
        [4,2,1,0],
        [3,8,5,1],
        [2,1,0,3],
]

print(jumpGame2D(points,0,0))
```

```python
def jumpGame2D(matrix): #can be made n*n by greedy
    nr = len(matrix)
    nc = len(matrix[0])
    
    dp = [[0]*(nc) for _ in range(nr+1)]
    
    for _ in range(nc):
        dp[nr][_] = 0
    
    for i in range(nr-1,-1,-1):
        for j in range(0,nc):
            ans = 0
            for k in range(nc):
                if i>0:
                    ans = max(matrix[i][k] -abs(k-j)+dp[i+1][k],ans)
                else:
                    ans = max(matrix[i][k] +dp[i+1][k],ans)
            dp[i][j] = ans
    
    return dp[0][0]
                
points = [
        [1,0,3,4],
        [4,2,1,0],
        [3,8,5,1],
        [2,1,0,3],
]

print(jumpGame2D(points))
```
