```python
def factorial(n:int) -> int:
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
print(factorial(5))  # Output: 120


def add(a: int|float, b: int|float) -> int|float:
    return a + b
print(add(3.0, 4))  # Output: 7


def stringAdd(*args:str) -> str:
    lst = []
    for arg in args:
        lst.append(arg.strip().capitalize())
    return ' '.join(lst)
print(stringAdd("  hello ", " world ", " this ", " is ", " python "))




lst2 : list[int] = [1, 2, 3, 4, 5]
tup2 : tuple[str, int, float] = ("example", 42, 3.14) #note difference between list and tuple
dict2 : dict[str, int] = {"one": 1, "two": 2, "three": 3}


def inArray(arr: list[int],target:int,isSorted: bool = False) -> bool:
    if not isSorted:
        return target in arr #linear search
    
    #binary search    
    s,e = 0, len(arr) - 1
    while s <= e:
        mid = (s + e) // 2
        if arr[mid] == target:
            return True
        elif arr[mid] < target:
            s = mid + 1
        else:
            e = mid - 1
    return False


print(inArray([1, 2, 3, 4, 5], 3))  # Output: True
print(inArray([1, 2, 3, 4, 5], 0))  # Output: False
print(inArray([5, 3, 1, 4, 2], 4))  # Output: True

```

```python
def ceil(arr,target):
    s,e = 0, len(arr) - 1
    ans = -1
    while s<=e:
        mid = (s + e) // 2
        if arr[mid] >= target:
            ans = mid
            e = mid - 1
        else:
            s = mid + 1
    return ans

arr = [1, 2, 8, 10, 10, 12, 19]
target = 5
print(ceil(arr, target))  # Output: 2 (index of element 8)

arr = [1, 2, 8, 10, 10, 12, 19]
target = 20
print(ceil(arr, target))  # Output: -1 (no element >= 20)


def findPeak(arr):
        s,e = 0, len(arr) - 1
        peak = -1
        
        while s<=e:
            mid = (s + e) // 2
            if (mid == 0 or arr[mid] > arr[mid - 1]) and (mid == len(arr) - 1 or arr[mid] > arr[mid + 1]):
                peak = mid
                break # important to break here to avoid infinite loop. We could also return mid here directly.
            elif mid > 0 and arr[mid] > arr[mid - 1]:
                s = mid + 1
            else:
                e = mid - 1
        return peak
    
arr1 = [1, 3, 20, 4, 1, 0, -1]
print(findPeak(arr1))  

arr2 = [1,2,3,4,5]
print(findPeak(arr2))  

arr3 = [0]
print(findPeak(arr3))
```
