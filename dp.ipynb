```python
def knapsack(prices,weights,n,capacity):
    if capacity == 0:
        return 0
    if n == 0:
        return 0
    ans = 0
    if weights[n-1]<=capacity:
        ans = max(prices[n-1]+knapsack(prices,weights,n-1,capacity-weights[n-1]),knapsack(prices,weights,n-1,capacity))
    else:
        ans = knapsack(prices,weights,n-1,capacity)
    
    return ans
```

```python
#reusing dp to reduce tc from n*total*total to n*total by reusing dp table
def minSplit(arr):
    n = len(arr)
    total = 0
    for _ in arr:
        total += _
    tar = total
    
    dp = [[False]*(tar+1) for _ in range(n+1)]
    
    for _ in range(0,n+1):
        dp[_][0] = True
        
    for i in range(1,n+1):
        for j in range(1,tar+1):
            if arr[i-1]<=j:
                dp[i][j] = dp[i-1][j-arr[i-1]] or dp[i-1][j]
            else:
                dp[i][j] = dp[i-1][j]
    
    ans = float("inf")   
    
    for i in range(1,total//2+1):
        p1 = i
        if dp[n][i] == False:
            continue
        p2 = total - p1
    
        possible_ans = abs(p1-p2)
        ans = min(ans,possible_ans)
    return ans

print(minSplit([1,2,4,5,8,10]))
```

```python
def LCS(a,b):
    n1,n2 = len(a),len(b)
    dp = [[0]*(n2+1) for _ in range(n1+1)]
    
    for i in range(1,n1+1):
        for j in range(1,n2+1):
            if a[i-1] == b[j-1]:
                dp[i][j] = 1 + dp[i-1][j-1]
            else:
                dp[i][j] = max(dp[i-1][j],dp[i][j-1])
    
    ans  = ""
    p1,p2 = n1,n2
    while(p1>0 and p2>0):
        if a[p1-1] == b[p2-1]:
            ans += a[p1-1]
            p1 -=1
            p2 -=1
        elif dp[p1-1][p2] > dp[p1][p2-1]:
            p1 -=1
        else:
            p2 -=1
            
    return dp[n1][n2],ans[::-1]

print(LCS("catip","matxop"))
print(LCS("caghgjvbfdfnjgauhlanjkntip","matfdhjgdsfhjbfshjdgkhjbxop"))
```

```python
def SCS(a,b):
    n1,n2 = len(a),len(b)
    dp = [[0]*(n2+1) for _ in range(n1+1)]
    
    for _ in range(0,n1+1):
        dp[_][0] = _
    for _ in range(0,n2+1):
        dp[0][_] = _
        
    for i in range(1,n1+1):
        for j in range(1,n2+1):
            if a[i-1] == b[j-1]:
                dp[i][j] = 1 + dp[i-1][j-1]
            else:
                dp[i][j] = min(1+dp[i-1][j],1+dp[i][j-1])
    
    ans  = ""
    p1,p2 = n1,n2
    while(p1>0 and p2>0):
        if a[p1-1] == b[p2-1]:
            ans += a[p1-1]
            p1 -=1
            p2 -=1
        elif dp[p1-1][p2] < dp[p1][p2-1]:
            ans += a[p1-1]
            p1 -=1
        else:
            ans += b[p2-1]
            p2 -=1
            
    while(p1>0):
            ans += a[p1-1]
            p1 = p1 - 1
    while(p2>0):
            ans += b[p2-1]
            p2 = p2 - 1

    return dp[n1][n2],ans[::-1]

print(SCS("geek","geekforgeeks"))
print(SCS("geek","eke"))
```

```python
def editdistance(a,b):
    n1 = len(a)
    n2 = len(b)
    
    dp = [[0]*(n2+1) for i in range(n1+1)]
    
    for x in range(n2+1):
        dp[0][x] = x
    for x in range(n1+1):
        dp[x][0] = x
    
    for i in range(1,n1+1):
        for j in range(1,n2+1):
            if a[i-1] == b[j-1]:
                dp[i][j] = dp[i-1][j-1] # pass / do nothing 
            else :
                dp[i][j] = min(dp[i-1][j]+1,dp[i][j-1]+1,dp[i-1][j-1]+1) # delete,insert,replace respectively
    
    return dp[n1][n2]

print(editdistance("horse","rose"))

# horse -> rorse -> rose
# rose -> hose -> horse  
```

```python
def wordBreak(words:set,target:str,i):
    if i==len(target):
        return 1
    ans = 0
    for j in range(i,len(target)):
        prefix = target[i:j+1]
        if prefix in words:
            ans += wordBreak(words,target,j+1)
    return ans

arr = set(["i","like","sam","sung","samsung","mob","mobile","ile"])
sen = "ilikesamsungmobile"

print(wordBreak(arr,sen,0))
```

```python
def wordBreak(words:set,target:str,i):
    if i==len(target):
        return [[]]
    ans = []
    for j in range(i,len(target)):
        prefix = target[i:j+1]
        if prefix in words:
            sub = wordBreak(words,target,j+1)
            ans += [[prefix] + l for l in sub]
    return ans

arr = set(["i","like","sam","sung","samsung","mob","mobile","ile"])
sen = "ilikesamsungmobile"

print(wordBreak(arr,sen,0))
```

```python
def jumpGame2D(matrix,r,p):
    if r == len(matrix):
        return 0
    ans = 0
    for i in range(len(matrix[0])):
        if r > 0:
            ans = max(ans,matrix[r][i] - abs(i-p) + jumpGame2D(matrix,r+1,i))
        else:
            ans = max(ans,matrix[r][i] + jumpGame2D(matrix,r+1,i))
    
    return ans
points = [
        [1,0,3,4],
        [4,2,1,0],
        [3,8,5,1],
        [2,1,0,3],
]

print(jumpGame2D(points,0,0))
```

```python
def jumpGame2D(matrix):
    nr = len(matrix)
    nc = len(matrix[0])
    
    dp = [[0]*(nc) for _ in range(nr+1)]
    
    for _ in range(nc):
        dp[nr][_] = 0
    
    for i in range(nr-1,-1,-1):
        for j in range(0,nc):
            ans = 0
            for k in range(nc):
                if i>0:
                    ans = max(matrix[i][k] -abs(k-j)+dp[i+1][k],ans)
                else:
                    ans = max(matrix[i][k] +dp[i+1][k],ans)
            dp[i][j] = ans
    
    return dp[0][0]
                
points = [
        [1,0,3,4],
        [4,2,1,0],
        [3,8,5,1],
        [2,1,0,3],
]

print(jumpGame2D(points))
```

```python
def houseRobber3(node,memo):
    
    if node is None:
        return 0
    if node in memo:
        return memo[node]
    a,b = 0,0
    if node.left:
        a = node.val + houseRobber3(node.left.left,memo)+houseRobber3(node.left.right,memo)
    if node.right:
        a += houseRobber3(node.right.left,memo) + houseRobber3(node.right.right,memo)
    b = houseRobber3(node.left,memo) + houseRobber3(node.right,memo)
    
    memo[node] =  max(a,b)
    return memo[node]
```

```python
def houseRobber3DFSDP(node):
    
    def dfs(node):
        if node is None:
            return (0,0) # like a tabulation martix row
        
        leftTake,leftSkip = dfs(node.left)
        rightTake,rightSkip = dfs(node.right)
        
        take = node.val + leftSkip + rightSkip
        skip = max(leftTake,leftSkip) + max(rightTake,rightSkip) # note
        
        return (take,skip)
    
    return dfs(node)


# Tree:
#         10
#        /  \
#       1    1
#      / \  / \
#    10 10 10 10

# Step by step:
# - Leaf nodes (value 10): rob = 10, skip = 0 → (10,0)
# - Intermediate nodes (value 1): children return (10,0) each
#     rob = 1 + 0 + 0 = 1
#     skip = max(10,0) + max(10,0) = 20
#     → (1,20)
# - Root node (value 10): children return (1,20) each
#     rob = 10 + 20 + 20 = 50
#     skip = max(1,20) + max(1,20) = 40   (Thats why max is important)
#     → (50,40)

# Final answer = max(50,40) = 50
```

```python
def height(node):
    if node == None:
        return 0
    
    a = height(node.left)
    b = height(node.right)
    return 1 + max(a,b)

def maxSumPath(node):
    if node == None:
        return 0
    
    a = maxSumPath(node.left)
    b = maxSumPath(node.right)
    return node.val + max(a,b,0)

def heightAndDiameter(node):
    if node == None:
        return (0,0)
    
    lh,ld = heightAndDiameter(node.left)
    rh,rd = heightAndDiameter(node.right)
    nh = 1 + max(lh,rh)
    nd = max(ld,rd,1+lh+rh)
    return (nh,nd)


def maxSumPathAndPathDiameter(node):
    if node == None:
        return (0,0)
    
    l1,l2 = maxSumPathAndPathDiameter(node.left)
    r1,r2 = maxSumPathAndPathDiameter(node.right)
    n1 = node.val + max(l1,r1,0)
    n2 = max(l2,r2,node.val+l1+r1) #dont put 0 guard here
    return (n1,n2)
```

```python
def NodeToLeafMaxSumPath(node):
    if node == None:
        return float("-inf")
    
    a = NodeToLeafMaxSumPath(node.left)
    b = NodeToLeafMaxSumPath(node.right)
    return node.val + max(a,b,0)
```

```python
def SCS(a,b):
    n1,n2 = len(a),len(b)
    dp = [[""]*(n2+1) for _ in range(n1+1)]
    
    for _ in range(0,n1+1):
        dp[_][0] = a[0:_]
    for _ in range(0,n2+1):
        dp[0][_] = b[0:_]
        
    for i in range(1,n1+1):
        for j in range(1,n2+1):
            if a[i-1] == b[j-1]:
                dp[i][j] = dp[i-1][j-1] + a[i-1]
            else:
                c1 = dp[i-1][j]
                c2 = dp[i][j-1]
                dp[i][j] = c1 + a[i-1] if len(c1)<len(c2) else c2 + b[j-1]
    
    return dp[n1][n2]

print(SCS("geek","geekforgeeks"))
print(SCS("geek","eke"))
```
