```python
class Heap():
    def __init__(self):
        self.arr = []
    
    def left(self,i):
        return 2*i
    def right(self,i):
        return 2*i + 1
    def parent(self,i):
        return i//2
    
    def add(self,val):
        self.arr.append(val)
        self.upheap(len(self.arr)-1)
    
    def pop(self):
        if len(self.arr)==0:
            raise Exception("popping from empty heap")

        out = self.arr[0]
        self.arr[0] = self.arr[-1]
        self.arr.pop()
        if self.arr:
            self.downheap(0)
        return out

    def swap(self,i,j):
        self.arr[i],self.arr[j] = self.arr[j],self.arr[i]
        
    def upheap(self,i):
        if i == 0:
            return
        parent = i//2
        if self.arr[i]<self.arr[parent]:
            self.swap(i,parent)
            self.upheap(parent)
    
    def downheap(self,p):
        if p>=len(self.arr):
            return
        l = self.left(p)
        r = self.right(p)
        smallerChild = l
        
        if r<len(self.arr):
            smallerChild = l if self.arr[l] < self.arr[r] else r
        elif l<len(self.arr):
            pass
        else:
            return
        
        if self.arr[smallerChild]<self.arr[p]:
            self.swap(smallerChild,p)
            self.downheap(smallerChild)


heap = Heap()

arr = [1,4,7,2,5,8,9,6,3]

for _ in arr:
    heap.add(_)

print(heap.arr)
while heap.arr:
    print(heap.pop())

print("\n")
for _ in arr:
    heap.add(-_)

while heap.arr:
    print(-heap.pop())
    
    
list
        
```

```python
def kLargest(arr,k):
    from heapq import heappop,heappush
    
    heap = []
    for _ in arr:
        heappush(heap,_)
        if len(heap) > k:
            heappop(heap)
    ans = []
    while heap:
        ans.append(heappop(heap))
    return ans

arr = [1,4,7,2,5,8,3,6,9]

print(kLargest(arr,3))
```
