```python
def powerOf4(n):
    return n == 1 or (n>0 and n%4==0 and powerOf4(n/4))

print(powerOf4(4))
print(powerOf4(5))
print(powerOf4(0))
print(powerOf4(1))
print(powerOf4(16))
print(powerOf4(64))
print(powerOf4(-64))
```

```python
def countways(n):
    def encode(s):
        j = 0
        total = []
        while j<len(s):
            c = s[j]
            count = 1
            while(j+1<len(s) and s[j]==s[j+1]):
                count += 1
                j += 1
            total.append(str(count)+c)
            j += 1
        
        return "".join(total)
    def main(n):
        if n == 1:
            return "1"
        return encode(main(n-1))
    
    return main(n)

print(countways(4))

```
