```python
import random, collections
class TreeNode():
    def __init__(self,val,left=None,right=None):
        self.val = val
        self.left = left
        self.right = right
    
    def __str__(self):
        return str(self.val)
    
    @staticmethod
    def generate(size):
        arr = collections.deque()
        for _ in range(size):
            arr.append(random.randint(-9,9))
        
        root = TreeNode(arr.popleft())
        
        for num in arr:
            root = TreeNode.insert(root,num)
        
        return root
    
    @staticmethod
    def insert(node,val):
        newNode = TreeNode(val)

        if node is None:
            return newNode
        
        head = node
        while node:
    
            if val < node.val:
                if node.left is None:
                    node.left = newNode
                    break
                node = node.left
            else:
                if node.right is None:
                    node.right = newNode
                    break
                node = node.right
        return head
    
    @staticmethod
    def display(node,step=0):
        if node is None:
            return
        TreeNode.display(node.left,step+1)
        if step ==0:
            print(node)
        else:
            print("     "*(step-1)+"|--->",node.val,sep="")
        
        TreeNode.display(node.right,step+1)
        

node = TreeNode.generate(10)

TreeNode.display(node)
```

```python
def serializePre(node,ans=None):
    if ans is None:
        ans = []
    if node is None:
        ans.append(None)
        return ans
    
    ans.append(node.val)
    serializePre(node.left,ans)
    serializePre(node.right,ans)
    return ans
    
    
root = TreeNode.generate(15)
TreeNode.display(root)

print(serializePre(root))
```

```python
def serializePre(node):
    ans = []
    stack = [node]
    while stack:
        out = stack.pop()
        ans.append(out.val if out else None)
        if out:
            stack.append(out.left)
            stack.append(out.right)
    return ans

root = TreeNode.generate(15)
TreeNode.display(root)

print(serializePre(root))
```

```python
def isBalanced(root):
    def helper(node):
        if node is None:
            return 0

        h1 = helper(node.left)
        if h1 <0:
            return -1
        h2 = helper(node.right)
        if h2 <0:
            return -1

        if abs(h1-h2)>1:
            return -1
        
        return 1 + max(h1,h2)
    
    ans = helper(root)
    return ans != -1
```
