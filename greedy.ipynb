```python
# wow
from functools import lru_cache
def naive1(arr,fuel,tar): # n*tar 
    
    n = len(arr)
    @lru_cache(None)
    def helper(i,reach):
        
        if reach >= tar:
            return 0
        if i == n:
            return float("inf")
        dist, fuel = arr[i]
        
        if dist<=reach:
            return min(1 + helper(i+1, (reach-dist)+dist+fuel),helper(i+1,reach))
        else:
            return helper(i+1,reach)
    
    return helper(0,fuel)
target = 120
start_fuel = 10
stations = [(10, 20), (15, 100)]
print(naive1(stations,start_fuel,target))

target = 100
start_fuel = 10
stations = [(10, 60), (20, 5), (25, 5), (30, 5)]
print(naive1(stations,start_fuel,target))


target = 200
start_fuel = 90
stations = [(10, 20), (90, 100),(191,10)]
print(naive1(stations,start_fuel,target))
```

```python
# wow
from functools import lru_cache
def naive2(arr,fuel,tar): # n*tar 
    
    n = len(arr)
    @lru_cache(None)
    def helper(i,reach):
        
        if reach >= tar:
            return 0
        if i == n:
            return float("inf")
        dist = arr[i]
        
        if dist<=reach:
            return min(1 + helper(i+1, dist+fuel),helper(i+1,reach))
        else:
            return helper(i+1,reach)
    
    return helper(0,fuel)
target = 120
start_fuel = 50
stations = [10, 15 , 25 , 40, 70]
print(naive2(stations,start_fuel,target))
```

```python
from heapq import heapify,heappop,heappush 
def joiningRopes(arr):
    heapify(arr)
    ans = 0
    while len(arr) >= 2:
        
        smallest = heappop(arr)
        second_smallest = heappop(arr)
        cost = smallest + second_smallest
        ans += cost
        heappush(arr,cost)
    return ans
        
```

```python
from heapq import heappush,heappop
def kSmallestPairs(nums1,nums2,k): # dijkstra like solution
    i,j,n1,n2 = 0,0,len(nums1),len(nums2)
    
    isVisited = {(i,j)}
    heap = [(nums1[i]+nums2[j],i,j)]
    
    ans = []
    
    while(heap and len(ans)<k):
        
        s,i,j = heappop(heap)
        ans.append([nums1[i],nums2[j]])
        
        c1,c2 = None,None
    
        if i+1<n1:
            c1 = (nums1[i+1]+nums2[j],i+1,j)
        if j+1<n2:
            c2 = (nums1[i]+nums2[j+1],i,j+1)
        
        if (c1 is not None) and (c1  not in isVisited):
            isVisited.add((c1[1],c1[2]))
            heappush(heap,c1)
        if (c2 is not None) and (c2  not in isVisited):
            isVisited.add((c2[1],c2[2]))
            heappush(heap,c2)
    return ans
            
```

```python
def twoSum(arr:list[int],tar:int) -> list[tuple[int,int]]:
    arr.sort()
    i,j = 0, len(arr)-1
    ans: list[tuple[int,int]] = []
    while(i<j):
        
        if i > 0 and arr[i] == arr[i-1]:
            i += 1
            continue
        # if j < len(arr)-1 and arr[j] == arr[j+1]: # either is fine. both checks not needed
        #     j -= 1
        #     continue
        sum = arr[i] + arr[j]
        
        if sum == tar:
            ans.append((arr[i],arr[j]))
            i += 1
            j -= 1
        elif sum < tar:
            i += 1
        else:
            j -=1
    
    return ans


arr = [1,4,4,2,2,9,5,5,7,8,2,3,6,6,6,4,4,1,2]

print(twoSum(arr,7))
```
