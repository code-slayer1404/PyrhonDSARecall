```python
from typing import Optional


class Node:
    def __init__(self, arr: Optional[list[Optional["Node"]]] = None, end: bool = False):
        self.arr: list[Optional["Node"]] = arr if arr is not None else [None] * 26
        self.end: bool = end


class Trie:
    def __init__(self):
        self.root: Node = Node()
        self.size: int = 0

    def insertIter(self, s: str) -> None:
        node: Node = self.root

        for ch in s:
            index = ord(ch) - ord("a")

            child = node.arr[index]
            if child is None:
                child = Node()
                node.arr[index] = child

            node = child

        if not node.end:
            node.end = True
            self.size += 1

    def search(self, s: str) -> bool:
        node: Node = self.root

        for ch in s:
            index = ord(ch) - ord("a")

            child = node.arr[index]
            if child is None:
                return False

            node = child

        return node.end

    def startsWith(self, s: str) -> bool:
        node: Node = self.root

        for ch in s:
            index = ord(ch) - ord("a")

            child = node.arr[index]
            if child is None:
                return False

            node = child

        return True

    def remove(self, s: str) -> bool:
        node: Node = self.root

        for ch in s:
            index = ord(ch) - ord("a")

            child = node.arr[index]
            if child is None:
                return False

            node = child

        if node.end:
            node.end = False
            self.size -= 1
            return True

        return False

    def printAllWords(self) -> None:
        stack: list[tuple[Node, str]] = [(self.root, "")]

        while stack:
            node, path = stack.pop()

            if node.end:
                print(path)

            for i in range(25, -1, -1):
                child = node.arr[i]
                if child is not None:
                    stack.append((child, path + chr(i + ord("a"))))
```
