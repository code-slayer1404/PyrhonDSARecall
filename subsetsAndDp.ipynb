```python
#a dp problem as ans only depends on length of p
def subsetsOfGivenLength(p, up, l):
    if l == 0:
        return 1
    if len(up) == 0:
        return 0
    
    left = subsetsOfGivenLength(p + up[0], up[1:], l - 1)
    right = subsetsOfGivenLength(p, up[1:], l)
    
    return left + right

print(subsetsOfGivenLength("", "cats", 3))  # Output: 1
print(subsetsOfGivenLength("", "I lovecats and dogs and most animals. I also love python and java", 5))  # Output: 1
```

```python
def longestCommonSubstring(s1,s2,l1=None,l2=None,pathLength=0):
    if l1 == None or l2 == None:
        l1,l2 = len(s1), len(s2)
    
    if l1 == 0 or l2 == 0:
        return pathLength
    
    ans = 0
    if s1[l1-1] == s2[l2-1]:
        ans = max(longestCommonSubstring(s1,s2,l1-1,l2-1,pathLength+1),longestCommonSubstring(s1,s2,l1-1,l2,pathLength),longestCommonSubstring(s1,s2,l1,l2-1,pathLength))
    else:
        ans = max(pathLength,longestCommonSubstring(s1,s2,l1-1,l2,0),longestCommonSubstring(s1,s2,l1,l2-1,0))
    return ans

print(longestCommonSubstring("supercar","car"))
print(longestCommonSubstring("memoization","tabulation"))
```

```python
def work(easy,hard):
    n = len(easy)
    memo = {}
    def helper(i,canDoHard=True):
        
        if i == n:
            return 0
        if (i,canDoHard) in memo:
            return memo[(i,canDoHard)]
        ans = None
        if canDoHard:
            memo[(i,canDoHard)] = max(0+helper(i+1,True),easy[i]+helper(i+1,False),hard[i]+helper(i+1,False))
        else:
            memo[(i,canDoHard)] = max(0+helper(i+1,True),easy[i]+helper(i+1,False))
        return memo[(i,canDoHard)]
    
    return helper(0,True)

easy_work = [10, 5, 20, 15, 8]
hard_work = [20, 30, 10, 40, 25]

print(work(easy_work,hard_work))
```

```python
def max_earnings_memoized(n, easy_work, hard_work):
    # memo[day][0]: max earnings from `day` onward, given we rested on `day-1`
    # memo[day][1]: max earnings from `day` onward, given we worked on `day-1`
    memo = [[-1 for _ in range(2)] for _ in range(n)]

    def solve(day, rested_yesterday):
        if day == n:
            return 0
        
        state = 0 if rested_yesterday else 1
        if memo[day][state] != -1:
            return memo[day][state]

        # Option 1: Do easy work today. We worked, so rested_yesterday = False
        res1 = easy_work[day] + solve(day + 1, False)

        # Option 2: Do no work today. We rested, so rested_yesterday = True
        res2 = solve(day + 1, True)
        
        result = max(res1, res2)
        
        # Option 3: Do hard work today, only if we rested yesterday.
        if rested_yesterday:
            res3 = hard_work[day] + solve(day + 1, False)
            result = max(result, res3)

        memo[day][state] = result
        return result

    # Start recursion on day 0, assuming we "rested" on day -1 to allow a hard task
    return solve(0, True)

# Example usage
n = 5
easy_work = [10, 5, 20, 15, 8]
hard_work = [20, 30, 10, 40, 25]

print(f"Maximum earnings: {max_earnings_memoized(n, easy_work, hard_work)}")
```

```python
def eggDrop(n,e):
    
    if e == 0:
        return n
    if n <= 1:
        return n
    
    ans  = float("inf")
    for k in range(0,n):
        temp = max(eggDrop(k-1,e-1),eggDrop(n-k,e))
        ans = min(ans,temp)
        
    return ans
```

```python
class Heap:
    def __init__(self,arr=None):
        self._arr = [] if arr is None else arr
        self.size:int = 0 if arr is None else len(arr)
        if arr is not None:
            self._heapify()
    
    def get_left(self,i:int):
        if 2*i+1 >= self.size:
            return None
        return 2*i+1
    
    def get_right(self,i:int):
        if 2*i+2 >= self.size:
            return None
        return 2*i+2
    
    def get_parent(self,i:int):
        return (i-1)//2
    
    def get_value(self,i):
        return self._arr[i]

    def swap(self,i,j):
        self._arr[i],self._arr[j] = self._arr[j],self._arr[i]
    
    def upheap(self,node):
        if node == 0 or node is None:
            return
        parent = self.get_parent(node)
        if self.get_value(node) < self.get_value(parent):
            self.swap(node,parent)
            self.upheap(parent)

    def downheap(self,node):  
        if node is None:
            return

        left = self.get_left(node)
        right = self.get_right(node)
        
        if left is None and right is None:
            return

        smaller = None
        if left is not None and right is not None:
            smaller = left if self.get_value(left)<=self.get_value(right) else right
        elif left is not None:
            smaller = left
        else:
            smaller = right
        if self.get_value(node)>self.get_value(smaller):    
            self.swap(node,smaller)
            self.downheap(smaller)

    def insert(self,val):
        self._arr.append(val)
        self.size = self.size + 1
        index = self.size - 1
        self.upheap(index)

    def pop(self):
        if self.size == 0:
            raise Exception("heap is empty")
        
        root = self.get_value(0)
        if self.size == 1:
            self._arr.pop()
            self.size = 0
            return root

        self._arr[0] = self._arr.pop()
        self.size = self.size - 1
        self.downheap(0)
        return root
    
    def peek(self):
        if self.size == 0:
            raise Exception("heap is empty")
        
        return self.get_value(0)
    
    def __str__(self) -> str:
        return str(self._arr) + " size = "+str(self.size)
    
    def _heapify(self):
        for _ in range(len(self._arr)):
            self.upheap(_)
        


from random import randint
arr = []
for _ in range(10):
    arr.append(randint(0,20))

minHeap = Heap(arr)

    
# print(minHeap.size)
print(minHeap)
while minHeap.size>0:
    print(minHeap.peek())
    minHeap.pop()
    print(minHeap)
```

```python
from collections import defaultdict,deque
def getAdjacencyList(arr):
    d = defaultdict(list)
    for (v1,v2) in arr:
        d[v1].append(v2)
        # d[v2].append(v1) # for undirected/bi-directional
    return d

def bfs(graph):
    al = getAdjacencyList(graph)
    isVisited = defaultdict(bool)
    
    source = graph[0][0]
    queue = deque()
    queue.append(source)
    isVisited[source] = True
    ans = []
    while queue:
        curr = queue.popleft()
        ans.append(curr)
        for neighbour in al[curr]:
            if not isVisited[neighbour]:
                isVisited[neighbour] = True
                queue.append(neighbour)
    return ans

arr = [(1,2),(1,5),(1,3),(2,3),(2,4),(2,5)]
arr2 = [(1,2),(1,3),(2,4),(2,5),(3,6),(3,7)]   

print(bfs(arr))
print(bfs(arr2))
```

```python
class Trie:
    class Node:
        def __init__(self):
            self.children = [None]*26  # Assuming lowercase English letters
            self.isEnd = False
    def __init__(self):
        self.root = self.Node()
        self.size = 0
    
    def insert(self,node,word,i):
        if i == len(word):
            if node.isEnd is False:
                self.size += 1
            node.isEnd = True
            return None
        
        char = word[i]
        index  = ord(char) - ord("a")
        if node.children[index] is None:
            node.children[index] = self.Node()

        self.insert(node.children[index],word,i+1)
    
    def find(self,node,word,i):
        if node is None:
            return False
        
        if i == len(word):
            return node.isEnd
        
        char = word[i]
        index = ord(char) - ord("a")
        
        if node.children[index] is None:
            return False
        
        return self.find(node.children[index],word,i+1)
    
    def startswith(self,node,word,i):
        if i == len(word):
            return True
        index = ord(word[i]) - ord("a")
        
        if node.children[index] is None:
            return False
        return self.startswith(node.children[index],word,i+1)
        
    
    
trie = Trie()

trie.insert(trie.root,"apple",0)
trie.insert(trie.root,"app",0)

print(trie.find(trie.root,"ap",0))
# print(trie.find(trie.root,"ball",0))
print(trie.find(trie.root,"apple",0))
print(trie.find(trie.root,"app",0))

print(trie.startswith(trie.root,"appm",0))

print(trie.size)
    
```

```python
def isUnique(arr,i):
    return i == 0 or arr[i-1] != arr[i]

def removeDup(arr):
    
    insertPointer = 0
    for readPointer in range(0,len(arr)):
        if isUnique(arr,readPointer):
            arr[insertPointer] = arr[readPointer]
            insertPointer += 1
            
    newLength = insertPointer
    for dup in range(insertPointer,len(arr)):
        arr[dup] = None
    return newLength    

arr = [1,1,2,3,4,4,4,5,6,6,7]
ans= removeDup(arr)
print(arr,ans)
```

```python
def isNotZero(arr,i):
    return arr[i] != 0

def removeZero(arr):
    
    insertPointer = 0
    for readPointer in range(0,len(arr)):
        if isNotZero(arr,readPointer):
            arr[insertPointer] = arr[readPointer]
            insertPointer += 1
            
    newLength = insertPointer
    for zero in range(insertPointer,len(arr)):
        arr[zero] = None
    return newLength    

arr = [0,1,1,0,2,3,0,4,4,4,5,6,0,6,7]
ans= removeZero(arr)
print(arr,ans)
```

```python
def Die(n,f,tar):
    
    dp = [[0]*(tar+1) for _ in range(n+1)]
    

    dp[0][0] = 1
        
    for i in range(1,n+1):
        for j in range(1,tar+1):
            for k in range(1,f+1):
                if k<=j:
                    dp[i][j] += dp[i-1][j-k]
            
    
    return dp[n][tar]


print(Die(30,6,150))
print(Die(7,6,16))
```

```python
def test(n,f,tar):
    if tar==0 and n==0:
        return 1
    if n == 0:
        return 0
    if f == 0:
        return 0
    ans = 0
    if f<=tar:
        ans += test(n-1,f,tar-f)
        ans += test(n,f-1,tar)
    else:
        ans += test(n,f-1,tar)
    
    return ans

print(test(7,6,16))
```
