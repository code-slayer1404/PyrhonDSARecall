```python
def doOverlap(interval1:tuple[int,int],interval2:tuple[int,int]) -> bool:
    smax = max(interval1[0],interval2[0])
    emin = min(interval1[1],interval2[1])
    return smax<=emin
def mergeIntervals(interval1:tuple[int,int],interval2:tuple[int,int]) -> tuple:
    if doOverlap(interval1,interval2) is False:
        # raise Exception("cannot be merged")
        return ()
    smin = min(interval1[0],interval2[0])
    emax = max(interval1[1],interval2[1])
    return (smin,emax)



print(doOverlap((1, 3), (3, 6)))       # True
print(mergeIntervals((1, 3), (3, 6)))  # (1, 6)
print(doOverlap((6, 9), (3, 6)))       # True
print(mergeIntervals((6, 9), (3, 6)))  # (3, 9)
print()

print(doOverlap((1, 2), (3, 4)))       # False
print(mergeIntervals((1, 2), (3, 4)))  # ()
print(doOverlap((5, 7), (8, 10)))      # False
print(mergeIntervals((5, 7), (8, 10))) # ()
print()

print(doOverlap((2, 6), (2, 6)))       # True
print(mergeIntervals((2, 6), (2, 6)))  # (2, 6)
```

```python
class Node:
    def __init__(self,end=False):
        self.children: dict = {}
        self.end: bool = end

class Trie:
    def __init__(self):
        self.root: Node = Node()
        self.size: int = 0
    
    def insert(self,s:str):
        
        i = 0
        node: Node = self.root
        while(i<len(s)):
            if s[i] not in node.children:
                node.children.update({s[i]:Node()})
            
            node = node.children[s[i]] 
            i += 1
        
        if node.end == False:
            node.end = True
            self.size += 1
            
    def search(self,s):
        
        i = 0
        node: Node = self.root
        while(i<len(s)):
            if node.children[s[i]] == None:
                return False
            
            node = node.children[s[i]]
            i += 1
        
        return node.end
    
    def startsWith(self,s:str):
        
        i = 0
        node: Node = self.root
        while(i<len(s)):
            if node.children[s[i]] == None:
                return False

            node = node.children[s[i]]
            i += 1
        
        return True
    
    def remove(self,s:str):
        
        i = 0
        node = self.root
        
        while(i<len(s)):
            if node.children[s[i]] == None: 
                return False
            node = node.children[s[i]] 
            i += 1
            
        if node.end == True: 
            node.end = False
            self.size -= 1
            return True
        
    # my doing inspired by dfs. correct in 1st attempt
    def printAllWords(self):        
        stack: list[tuple[Node,str]] = [(self.root,"")]
        
        while(stack):
            node,path = stack.pop()
    
            if  node.end == True:
                print(path)
            for childKey in node.children:
                child = node.children[childKey]
                stack.append((child,path+childKey))
                

t = Trie()

t.insert("cat")
t.insert("ca")
t.insert("catto")
t.printAllWords()
print()

print(t.search("cat"))
print(t.search("c"))
print()

t.remove("cat")
t.insert("dog")
t.printAllWords()
print()

print(t.startsWith("ca"))
print()


print(t.size)
print()

t.printAllWords()
```

```python
def subsetStack(s):
    
    n = len(s)
    stack = [("",0)]
    ans = []
    
    while(stack):
        p,i = stack.pop()
        
        if i == n:   # base case
            ans.append(p)
            continue    # important. or the below 2 lines should be under 
                        # condition: i<n
            
        stack.append((p+s[i],i+1))
        stack.append((p,i+1))
    return ans


print(subsetStack("cat"))
```

```python
class TreeNode:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

    def __str__(self):
        return str(self.val)


class Tree:
    def __init__(self, root=None):
        self.root = root
        self.size = 0

    def create(self):
        self.size = 0
        val = int(input("Enter root: "))
        self.root = TreeNode(val)
        self.size += 1
        self._create(self.root)

    def _create(self, node):
        while True:
            choice = int(input(
                f"Press 1 to insert left of {node}, "
                f"2 to insert right of {node}, "
                f"0 to abort: "
            ))

            if choice == 0:
                break
            elif choice == 1:
                if node.left is None:
                    val = int(input("Enter value to be entered: "))
                    node.left = TreeNode(val)
                    self.size += 1
                self._create(node.left)
            elif choice == 2:
                if node.right is None:
                    val = int(input("Enter value to be entered: "))
                    node.right = TreeNode(val)
                    self.size += 1
                self._create(node.right)
            else:
                print("Invalid choice, try again.")


# Example usage
testTree = Tree()
testTree.create()
print("Tree size:", testTree.size)
```

```python
def allWindowsWithSumK(arr,k): # fails if 0s or negatives are there
    
    i,j,n,total = 0,0,len(arr),0
    ans = 0
    while(j<n):
        total += arr[j]
        
        while total>k and i<=j:
            total -= arr[i]
            i += 1
        if total == k:   
            ans += 1
            # print(arr[i:j+1])
        j += 1
    
    return ans
        
print(allWindowsWithSumK([1,2,5,5,2,8,10],10))
# print(allWindowsWithSumK([1,2,0,5,0,5,0,2,8,10,0],10)) fail


def allWindowsWithSumLessThanOrEqualToK(arr,k):
    
    i,j,n,total = 0,0,len(arr),0
    ans = 0
    while(j<n):
        total += arr[j]
        
        while total>k and i<=j:
            total -= arr[i]
            i += 1
        ans += j-i+1    # only works for <= problems and not == problems.
                        # if [i...j] is valid, [i+1..j],[i+2..j]...[j] are all valid
                        # special property of <= problems
        j += 1
    
    return ans
        
print(allWindowsWithSumLessThanOrEqualToK([1,2,5,5,2,8,10],10))
print(allWindowsWithSumLessThanOrEqualToK([1,2,0,5,0,5,0,2,8,10,0],10))


def allWindowsWithSumK0sAllowed(arr,k):
    
    return allWindowsWithSumLessThanOrEqualToK(arr,k) - allWindowsWithSumLessThanOrEqualToK(arr,k-1)
        
print(allWindowsWithSumK0sAllowed([1,2,0,5,0,5,0,2,8,10,0],10))
print(allWindowsWithSumK0sAllowed([1,2,0,5],7))
```

```python
# given arr[i] >= 0
def smallestWindowWithSumK(arr,k):
    
    i,j,n,total = 0,0,len(arr),0
    ans = float("inf")
    while(j<n):
        total += arr[j]
        
        while total>k and i<=j:
            total -= arr[i]
            i += 1
        if total == k:
            ans = min(j-i+1,ans)
        j += 1
    
    return ans
        
print(smallestWindowWithSumK([1,2,5,5,2,8,10],10))

def largestWindowWithSumK(arr,k):
    
    i,j,n,total = 0,0,len(arr),0
    ans = float("-inf")
    while(j<n):
        total += arr[j]
        
        while total>k and i<=j:
            total -= arr[i]
            i += 1
        if total == k:
            ans = max(j-i+1,ans)
        j += 1
    
    return ans
        
print(largestWindowWithSumK([1,2,5,5,2,8,10],10))
```
