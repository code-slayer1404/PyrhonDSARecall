```python
# final foolproof ideal
from random import randint
class Node():
    def __init__(self,val,next=None):
        self.val = val
        self.next = next
    def __str__(self):
        return str(self.val)

    @staticmethod
    def generate(length=10,minimum=-10,maximum=10):
        dummyHead = tail = Node(None)
        for _ in range(length):
            tail.next = Node(randint(minimum,maximum)) 
            tail = tail.next
        
        return dummyHead.next

    @staticmethod
    def display(head,description=""):
        temp = head
        while temp is not None:
            print(temp.val,end=" ")
            temp = temp.next
        print("\t\t",description)
    
    @staticmethod
    def reverse(head):
        prev = None
        curr = head
        
        while(curr is not None):
            next = curr.next
            curr.next = prev
            prev = curr
            curr = next
            
        return prev
    
    @staticmethod
    def getMid(head):
        slow = head
        fast = head
        
        while(fast is not None and fast.next is not None and fast.next.next is not None):
            slow = slow.next
            fast = fast.next.next
        
        return slow
    
    @staticmethod
    def splitFromMid(head):
        mid = Node.getMid(head)
        headOfNextHalf = mid.next
        mid.next = None
        return (head,headOfNextHalf)
    
    @staticmethod
    def mergeTwoSortedLinkedList(head1,head2):
        dummyhead = Node(None)
        insertPointer = dummyhead
        
        while(head1 is not None and head2 is not None):
            if head1.val < head2.val:
                insertPointer.next = head1
                head1 = head1.next
            else:
                insertPointer.next = head2
                head2 = head2.next
                
            insertPointer = insertPointer.next
        
        insertPointer.next = head1 if head1 is not None else head2
        
        return dummyhead.next
    
    @staticmethod
    def mergesort(head):
        if head is None or head.next is None:
            return head
        
        h1,h2 = Node.splitFromMid(head)
        m1,m2 = Node.mergesort(h1),Node.mergesort(h2)
        
        return Node.mergeTwoSortedLinkedList(m1,m2)
    
    @staticmethod
    def removeDup(head):
        prev = None
        curr = head
        dummyHead = Node(None)
        insertPointer = dummyHead
        
        while(curr is not None):
            if prev is None or curr.val != prev.val:
                insertPointer.next = curr
                insertPointer = insertPointer.next
            prev = curr
            curr = curr.next
        
        return dummyHead.next
        
    
    # @staticmethod
    # def reverseRecursiveStandard(prev,curr):
    #     # may replace commented lines with the comments
    #     if curr is None: #curr.next is None
    #         # curr.next = prev
    #         return prev #curr
    #     tail = Node.reverseRecursiveStandard(curr,curr.next)
        
    #     curr.next = prev
    #     return tail
    
    # @staticmethod
    # def reverseX(node):
    #     if node.next is None:
    #         return node
    #     temp = Node.reverseX(node.next)
    #     node.next.next = node
    #     node.next = None
    #     return temp
    
    @staticmethod
    def detectCycles(head):
        slow,fast = head,head
        
        meetingPoint = None
        while(fast is not None and fast.next is not None):
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                meetingPoint = slow
                break
            
        if meetingPoint is None:
            return None
        
        while head != meetingPoint:
            head = head.next
            meetingPoint = meetingPoint.next
            
        return meetingPoint
    # ds = x1 + x2, df = x1 + 2x2 + x3
    # df = 2ds => x1 + 2x2 + x3 = 2x1 + 2x2
    # => x1 = x3 (hence proved)
    
    # if list if perfectly circular
    # x1 = 0 => x3 = 0 => meeting point coicides with entry point i.e  also head


linkedList = Node.generate()
Node.display(linkedList,"list1")

reversedList = Node.reverse(linkedList)
Node.display(reversedList,"reverse of list1")


# mid = Node.getMid(reversedList)
# Node.display(mid)

part1,part2 = Node.splitFromMid(reversedList)
Node.display(part1,"first part of reverse of list1")
Node.display(part2,"second part of reverse of list1")

linkedList2 = Node.generate()
Node.display(linkedList2,"list2")
sortedList = Node.mergesort(linkedList2)

Node.display(sortedList,"sorted list2")
Node.display(Node.removeDup(sortedList),"sorted list2 without duplicates")


# print("recursive ones below")
# linkedList3 = Node.generate()
# Node.display(linkedList3)
# reversedList = Node.reverseRecursiveStandard(None,linkedList3)
# # reversedList = Node.reverseX(linkedList3)
# Node.display(reversedList)
```

```python
# a modified reverse to return both head and tail of reversed segment

def reverse(head:None|Node) -> tuple:
    prev: None|Node = None
    curr: None|Node = head
    
    while(curr is not None):
        next = curr.next
        curr.next = prev
        prev = curr
        curr = next
        
    return (prev,head)


def removeNodesInGroupOfK(head,k):
    
    dummyHead = Node(None,head)
    prev : Node = dummyHead
    curr = head
    while curr is not None:
        groupEnd = curr
        for _ in range(k-1):
            groupEnd = groupEnd.next
            if groupEnd is None:
                return dummyHead.next # or break
            
        nextHead = groupEnd.next
        groupStart = prev.next
        
        # cuttting
        prev.next = None # not necessary
        groupEnd.next = None # necessary
        
        headOfSegmentAfterReversal,tailOfSegmentAfterReversal = reverse(groupStart)
        
        # re-attaching
        prev.next = headOfSegmentAfterReversal
        tailOfSegmentAfterReversal.next = nextHead
        
        prev = tailOfSegmentAfterReversal
        curr = nextHead


    return dummyHead.next


linkedList3 = Node.generate(10)
Node.display(linkedList3)
ans  = removeNodesInGroupOfK(linkedList3,3)
Node.display(ans)
            
```

```python
def reverse(node):

    if node is None:
        return None

    if node.next is None:
        return node

    tail = reverse(node.next)
    node.next.next = node
    node.next = None
    return tail


l12 = Node.generate(5)

Node.display(l12)

l21 = reverse(l12)

Node.display(l21)
```

```python
# very good
def removeVal(head,val):
    if head is None:
        return None
    
    if head.val == val:
        return removeVal(head.next,val)
    else:
        head.next = removeVal(head.next,val)
    
    return head

l1 = Node.generate(40,0,6)
Node.display(l1)
l2 = removeVal(l1,5)
Node.display(l2)


def removeValIter(head,val):
    dummyHead = tail = Node(420)
    while head:
        if head.val != val:
            tail.next = head
            tail = tail.next
        head = head.next
        
    tail.next = None # important
    return dummyHead.next

l1 = Node.generate(40,0,6)
Node.display(l1)
l2 = removeValIter(l1,5)
Node.display(l2)
```

```python
# goated
def addTwoNumbers2(l1,l2):
    stack1 = []
    stack2 = []
    
    node = l1
    
    while node:
        stack1.append(node.val)
        node = node.next
    node = l2
    while node:
        stack2.append(node.val)
        node = node.next
        
    carry = 0
    prev = None
    while stack1 or stack2 or carry:
        a = stack1.pop() if stack1 else 0
        b = stack2.pop() if stack2 else 0
        
        total = a+b+carry
        
        carry = total // 10
        digit = total % 10
        
        node = Node(digit)
        node.next = prev
        prev = node
    
    return node

l1 = Node.generate(18,0,9)
Node.display(l1)
l2 = Node.generate(18,0,9)
Node.display(l2)

l3 = addTwoNumbers2(l1,l2)
Node.display(l3)


l1 = Node.generate(999,0,9) # a very very large number 10^999. There are only 10^80 atoms in OU
Node.display(l1)
l2 = Node.generate(999,0,9)
Node.display(l2)

l3 = addTwoNumbers2(l1,l2)
Node.display(l3)
```
