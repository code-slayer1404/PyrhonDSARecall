For every prefix value `p = P[i]`, multiple indices `i` may exist.  

The question is not:  

**‚ÄúShould I store this prefix or not?‚Äù**  

The real question is:  

**‚ÄúAmong all valid indices for this prefix value, which one helps my objective?‚Äù**  

That single question determines **overwrite vs never overwrite**.  

---

### Reframing all three with one principle

Let the window be `[i, j]`:



\[
\text{length} = (j + 1) - i
\]



---

#### üîπ COUNT
- Every valid `i` contributes  
- We keep **all occurrences**  
- **Implementation:** store frequency  

---

#### üîπ LONGEST
- Want length as **large as possible**  
- That means `i` should be as **small as possible**  
- We keep the **earliest i**  
- **Implementation:** never overwrite  

---

#### üîπ SHORTEST
- Want length as **small as possible**  
- That means `i` should be as **large as possible**  
- We keep the **latest i**  
- **Implementation:** always overwrite  

</span>


##### sum[i,j] = sum[0,j] - sum(0,i-1), window length = j - i + 1
##### for 0 based prefix sum model:
sum[i,j] = P[j+1](sum of 0 to j elements) - P[i] (sum of 0 to i-1 elements), window length = j - i + 1. For all P[j+1]s check if P[i] (P[j+1]-k) exists.
##### for -1 based prefix sum model:
sum[i,j] = P[j](sum of 0 to j elements) - P[i-1] (sum of 0 to i-1 elements), window length = j - i + 1 (same). For all P[j]s check if P[i-1] (P[j]-k) exists.
##### Baseline to always remember:
sum of elements in window [i,j] = sum of elements till index j - sum of elements till index i-1 always. (window length is same j-i+1 here
) (index mapped to first j elements sum - index mapped to first i-1 elements sum)


```python
def test(arr,k):
    n = len(arr)
    
    hashmap = {0:1}
    curr = 0
    ans = 0
    
    for j in range(0,n):
        curr += arr[j] # P(j+1) (sum of first j elements)
        
        if curr - k in hashmap:
            ans += hashmap[curr-k]
        
        hashmap[curr] = hashmap.get(curr,0) + 1
        
    return ans

print(test([1,4,-2,3,5,-6,2,8],5))
print(test([1, 4, -2, 3],5))
        
```

```python

def test(arr, k):
    hashmap = {0: 0}      # P[0] = 0 occurs at index 0
    curr = 0              # curr == P[j+1]
    ans = 0

    for j in range(len(arr)):
        curr += arr[j]    # compute P[j+1] (sum of first j elements)

        # check if there exists i such that P[i] = P[j+1] - k
        if curr - k in hashmap:
            i = hashmap[curr - k] # sum of first i-1 elements
            # length = index mapped to sum of j elements - index mapped to sum of i-1 elements
            ans = max(ans, (j + 1) - i)

        # store earliest occurrence only
        if curr not in hashmap: ## vvip as we want longest window
            hashmap[curr] = j + 1 # sum of first j mapped at j+1 

    return ans


print(test([1,4,-2,3,5,-6,2,8],5))
print(test([1, 4, -2, 3],5))
print(test([1, 2, 3, 4, 5],15))
        
```

```python

def test(arr, k):
    hashmap = {0: 0}      # P[0] = 0 occurs at index 0
    curr = 0              # curr == P[j+1]
    ans = float("inf")

    for j in range(len(arr)):
        curr += arr[j]    # compute P[j+1] (sum of first j elements)

        # check if there exists i such that P[i] = P[j+1] - k
        if curr - k in hashmap:
            i = hashmap[curr - k] # sum of first i-1 elements
            ans = min(ans, (j + 1) - i)
            
        # ALWAYS overwrite to keep latest index
        hashmap[curr] = j + 1  # sum of first j mapped at j+1 

    return ans


print(test([1,4,-2,3,5,-6,2,8],5))
print(test([1, 4, -2, 3],5))
print(test([1, 2, 3, 4, 5],15))
print(test([1, 2, 3, 4, 5],16))
        
```

```python

def test(arr, k):
    hashmap = {0: -1}      # P is zero based here
    curr = 0              # curr == P[j]
    ans = 0

    for j in range(len(arr)):
        curr += arr[j]    # compute P[j] (sum of first j elements)

        # check if there exists i -1 such that P[i-1] = P[j] - k
        if curr - k in hashmap:
            imin1 = hashmap[curr - k] # sum of first i-1 elements
            # length = index mapped to sum of j elements - index mapped to sum of i-1 elements
            ans = max(ans, j -imin1)

        # store earliest occurrence only
        if curr not in hashmap: 
            hashmap[curr] = j # sum of first j mapped at j

    return ans


print(test([1,4,-2,3,5,-6,2,8],5))
print(test([1, 4, -2, 3],5))
print(test([1, 2, 3, 4, 5],15))
print(test([1, 2, 3, 4, 5],1))
        
```
