```python
def pivot(arr):
    
    s,e = 0, len(arr) -1
    ans = -1
    while(s<=e):
        
        m = s + (e-s)//2
        
        if m+1 < len(arr) and arr[m+1]<arr[m]:
            return m
        if arr[s]<arr[m]:
            if ans != -1:
                ans = m if arr[m]>arr[ans] else ans
            else:
                ans = m
            s = m + 1
        else:
            if ans != -1:
                ans = e if arr[e]>arr[ans] else ans
            else:
                ans = e
            e = m - 1      
    return ans


arr = [7,8,9,1,2,3,4]
print(pivot(arr))

arr = [4,6,1,2,3]
print(pivot(arr))

arr = [1,2,3,4,5]
print(pivot(arr))

arr = [4,5,6,1,2]
print(pivot(arr))

arr = [5,1,2,3,4]
print(pivot(arr))

```
