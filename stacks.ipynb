````python
#leetcode
def discount(prices):
    ans = [len(prices)]*len(prices)
    stack = []
    
    for i in range(len(prices)):
        
        while(stack and prices[i] < prices[stack[-1]]):
            out = stack.pop()
            ans[out] = i
            
        stack.append(i)
    
    for i in range(len(prices)):
        j = ans[i]
        if j<len(prices):
            prices[i] = prices[i] - prices[j]
    
    return prices

print(discount([1,8,5,4,4,2,7]))
````

````python
def addTwoNumbers(s1,s2):
    n1 = len(s1)
    n2 = len(s2)
    # for linked list, replace n1,n2 with stacks of s1 and s2
    i,j,c = n1-1,n2-1,0

    stack = []
    while i>-1 or j>-1 or c>0:
        a = s1[i] if i>-1 else 0
        b = s2[j] if j>-1 else 0
        
        total = int(a)+int(b)+c
        
        c = total // 10
        digit = total % 10
        stack.append(str(digit))
        i -= 1
        j -= 1
    return "".join(stack[::-1])

print(addTwoNumbers("91","9"))
print(addTwoNumbers("67","83"))
print(addTwoNumbers("9","9"))
print(addTwoNumbers("9999999999999","9546543525454154"))
print(9999999999999+9546543525454154)
````

````python
def simplifyPath(path):
    stack = []

    for part in path.split('/'):
        if part == '' or part == '.':
            continue
        elif part == '..':
            if stack:
                stack.pop()
        else:
            stack.append(part)

    return '/' + '/'.join(stack)
````

````python
def calculate(s):
    stack = []
    res = 0
    num = 0
    sign = 1

    for ch in s:
        if ch.isdigit():
            num = num * 10 + int(ch)

        elif ch in '+-':
            res += sign * num
            num = 0
            sign = 1 if ch == '+' else -1

        elif ch == '(':
            stack.append(res)
            stack.append(sign)
            res = 0
            sign = 1

        elif ch == ')':
            res += sign * num
            num = 0
            prev_sign = stack.pop()
            prev_res = stack.pop()
            res = prev_res + prev_sign * res

    return res + sign * num

print(calculate("5+5-2"))
````

````python
def scoreOfParentheses(s):
    stack = []
    active = 0
    
    for c in s:
        if c == "(":
            stack.append(active)
            active = 0
        else:
            paused = stack.pop()
            innerVal = max(1,2*active)
            active = paused + innerVal
    return active

print(scoreOfParentheses("()"))
print(scoreOfParentheses("(())"))
print(scoreOfParentheses("((()))"))
print(scoreOfParentheses("(())()"))
print(scoreOfParentheses("((())())"))
print(scoreOfParentheses("()()"))
print(scoreOfParentheses("()()()"))
````
