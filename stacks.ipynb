```python
def nextSmallerStrictAndpreviousSmallerNonStrict(arr):
    ans = [len(arr)]*len(arr)
    ans2 = [-1]*len(arr)
    
    stack = []
    
    for i in range(len(arr)):
        
        while(stack and arr[i] < arr[stack[-1]]):
        
            out = stack.pop()
            ans[out] = i
            
        if stack:
            ans2[i] = stack[-1]
        stack.append(i)
    
    return ans,ans2

print(nextSmallerStrictAndpreviousSmallerNonStrict([1,8,5,4,4,2,7]))
print(nextSmallerStrictAndpreviousSmallerNonStrict([1,8,5,4,4,2,7][::-1]))# subtract ans from len(arr) - 1
```

```python
# big brain
def nextSmallerStrict(arr):
    ans = [len(arr)]*len(arr)
    stack = []
    
    for i in range(len(arr)):
        
        while(stack and arr[i] < arr[stack[-1]]):
            out = stack.pop()
            ans[out] = i
            
        stack.append(i)
    
    return ans

print(nextSmallerStrict([1,8,5,4,4,2,7]))
temp = nextSmallerStrict([1,8,5,4,4,2,7][::-1])[::-1]

ans = [ len(temp)-1-i for i in temp]
print(ans) # prev smaller strict
```
