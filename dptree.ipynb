```python
def height(node):
    if node == None:
        return 0
    
    a = height(node.left)
    b = height(node.right)
    return 1 + max(a,b)

def maxSumPath(node):
    if node == None:
        return 0
    
    a = maxSumPath(node.left)
    b = maxSumPath(node.right)
    return node.val + max(a,b,0)

def heightAndDiameter(node):
    if node == None:
        return (0,0)
    
    lh,ld = heightAndDiameter(node.left)
    rh,rd = heightAndDiameter(node.right)
    nh = 1 + max(lh,rh)
    nd = max(ld,rd,1+lh+rh)
    return (nh,nd)


def maxSumPathAndPathDiameter(node):
    if node == None:
        return (0,0)
    
    l1,l2 = maxSumPathAndPathDiameter(node.left)
    r1,r2 = maxSumPathAndPathDiameter(node.right)
    n1 = node.val + max(l1,r1,0)
    n2 = max(l2,r2,node.val+l1+r1) #dont put 0 guard here
    return (n1,n2)
```

```python
def houseRobber3DFSDP(node):
    
    def dfs(node):
        if node is None:
            return (0,0) # like a tabulation martix row
        
        leftTake,leftSkip = dfs(node.left)
        rightTake,rightSkip = dfs(node.right)
        
        take = node.val + leftSkip + rightSkip
        skip = max(leftTake,leftSkip) + max(rightTake,rightSkip) # note
        
        return (take,skip)
    
    return dfs(node)
```

```python
def pathSum(node,tar):
    
    if node is None:
        return False
    
    if node.left == None and node.right == None:
        return tar == node.val
    
    a = pathSum(node.left,tar-node.val)
    
    if a == True:
        return True
    return pathSum(node.right,tar-node.val)
```

```python
def pathSumAny(node,tar):
    
    if node is None:
        return False

    if node.val == tar: # we dont do first invalid case as there are two for a tree node. we do last valid
        return True
    
    a = pathSumAny(node.left,tar-node.val)
    
    if a == True:
        return True
    return pathSumAny(node.right,tar-node.val)
```

```python
def pathSum2(node,tar,path):
    
    if node is None:
        return
    
    if node.left == None and node.right == None:
        if tar == node.val:
            path.append(node)
            print(path)
            path.pop()
        return
    
    path.append(node)
    pathSum2(node.left,tar-node.val,path)
    # path.pop()
    
    # path.append(node)
    pathSum2(node.right,tar-node.val,path)
    path.pop()
```

```python
#VVIP
def pathSumAny3(node,tar):
    
    if node is None:
        return 0

    ans = 0
    
    if node.val == tar: # we dont do first invalid case as there are two for a tree node. we do last valid
        ans = 1
    
    ans += pathSumAny3(node.left,tar-node.val)
    ans += pathSumAny3(node.right,tar-node.val)
    
    return ans
```
